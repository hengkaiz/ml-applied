test: using knnimputer to calculate nan values

```python
temp_df = df.copy()
temp_df = temp_df.drop(columns=["enrollee_id", "target"])
temp_df = temp_df.astype("category")
```

```python
temp_df.head()
```

```python
all_cols = temp_df.columns
for i in all_cols:
    temp_df[i] = temp_df[i].cat.codes
    
temp_df.head()
```

```python
temp_df = temp_df.replace(-1, np.nan)
```

```python
imputer = KNNImputer(n_neighbors=5, weights="uniform", metric="nan_euclidean")
```

```python
imputer.fit(temp_df)
temp_fill = imputer.transform(temp_df)
```

```python
temp_df = pd.DataFrame(temp_fill, columns=all_cols)
temp_df.head()
```

```python
temp_df = temp_df.round()
temp_df.head()
```

```python
df[~(df["company_size"].isnull()) & (df["company_type"].isnull())]
```

```python
df.groupby(["city_development_index", "company_type"])["company_size"].value_counts(dropna=False)
```

```python
# Bin city_development_index by 0.05 intervals
min_cdi = df["city_development_index"].min()
max_cdi = df["city_development_index"].max()
step = 0.05

bins = np.arange(min_cdi, max_cdi, step)

df["cdi_bin"] = pd.cut(df["city_development_index"], bins=bins)
```

```python
gender_df = df.groupby(["city_development_index", "city", "relevent_experience"])["gender"].value_counts(dropna=False).reset_index(name="gender_cnt")
```
